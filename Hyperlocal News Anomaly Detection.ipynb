```python
# Extracting Location from Article manually
import re

def extract_city(article):
    article = str(article).upper().strip()  
    match = re.match(r"(?:STRONG>)?([A-Z/\s]+):", article)
    if match:
        return match.group(1).strip()
    else:
        return None
                  
df['Header_Location'] = df['Article'].apply(extract_city)
```

```python
df[df['Header_Location'].isna()]
```

```python
# Cleaning the Text
import re
def clean_text(text):
    text = str(text)
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()
```

```python
df['Article'] = df['Article'].apply(clean_text)
df['Heading'] = df['Heading'].apply(clean_text)
```

```python
# NER Location Extraction

import spacy
nlp = spacy.load("en_core_web_sm")
def Extract_NER_Location(text):
    doc = nlp(text)
    return list(set(ent.text for ent in doc.ents if ent.label_ in ['GPE','LOC']))

df['NER_Location'] = df['Article'].apply(Extract_NER_Location)   
```

```python
# Normalize text for NER_Location Column and removing the repeated names of the location
def clean_location(locations):
    return list(set(loc.lower().strip() for loc in locations))
df['NER_Location'] = df['NER_Location'].apply(clean_location)
```

```python
# SENTIMENT ANALYSIS TO FIND EMOTION PATTERN
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

def get_Sentiments_scores(text):
    scores = sia.polarity_scores(text)
    return scores['compound']
```

```python
# Applying Sentiment to all our Articles
df['sentiment']= df['Article'].apply(get_Sentiments_scores)
```

```python
df.head(1)
```

```python
# Converting Sentiment numerics to Labels
def sentiment_label(score):
    if score >= 0.05:
        return "Positive"
    elif score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

df['sentiment_label'] = df['sentiment'].apply(sentiment_label)
df.head(3)
```

```python
# Topic Modeling
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
nltk.download('stopwords')

# Convert text into Numbers
stopwords = stopwords.words('english')
vector = CountVectorizer(max_df=0.9,min_df=5,stop_words=stopwords)
dtm = vector.fit_transform(df['Article'])
```

```python
# Built topic model
from sklearn.decomposition import LatentDirichletAllocation
Lda = LatentDirichletAllocation(n_components=5,random_state=42)
Lda.fit(dtm)
```

```python
words = vector.get_feature_names_out()

for i, topic in enumerate(Lda.components_):
    print(f"\nTopic {i+1}:")
    print([words[j] for j in topic.argsort()[-10:]])
```

```python
# Assigning Topic to Each Article
topic_values = Lda.transform(dtm)
df['Topic'] = topic_values.argmax(axis=1)
```

```python
df.head(2)
```

```python
# Converting Date column to Date_Time
df['Date'] = pd.to_datetime(df['Date'])
df['DayOftheWeek'] = df['Date'].dt.day_name()
df['month'] = df['Date'].dt.month
```

```python
df.head(1)
```

```python
# Anomaly Detection
# Converting Text to Labels
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['NewsType_Encoded'] = le.fit_transform(df['NewsType'])
df['sentiment_label'] = le.fit_transform(df['sentiment_label'])
```

```python
numeric_features = df[['sentiment','Topic','NewsType_Encoded','sentiment_label']]
```

```python
# Train Anomaly Detection Model (Isolation Forest)
from sklearn.ensemble import IsolationForest

iso = IsolationForest(
    n_estimators=100,
    contamination=0.05,   
    random_state=42
)

df['anomaly_score'] = iso.fit_predict(numeric_features)

# Interpret anomaly result
df['anomaly_label'] = df['anomaly_score'].apply(
    lambda x: 'Anomaly' if x == -1 else 'Normal'
)

```

```python
df['Header_Location'] = df['Header_Location'].str.strip()
df['Header_Location'] = df['Header_Location'].str.lower()
```

```python

# SOURCE DISCREPANCY DETECTION 
# Logistic Regression 

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


df = df.copy()
df = df.dropna(subset=['Article', 'Header_Location'])
df.reset_index(drop=True, inplace=True)


location_counts = df['Header_Location'].value_counts()
valid_locations = location_counts[location_counts >= 2].index

df = df[df['Header_Location'].isin(valid_locations)].copy()
df.reset_index(drop=True, inplace=True)

print("Remaining Locations:", df['Header_Location'].nunique())

# Encoding
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(df['Header_Location'])

# Convert Text to Vector

vectorizer = TfidfVectorizer(max_features=5000)
X = vectorizer.fit_transform(df['Article'])

# Train-Test Split 


X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)

#  Train Logistic Regression Model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

#  Evaluate Model
y_pred_test = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred_test)
print("Model Accuracy:", accuracy)


y_pred_full = model.predict(X)
y_prob_full = model.predict_proba(X)

df['Predicted_Location'] = encoder.inverse_transform(y_pred_full)


# Create Boolean Mismatch Column
df['Source_Mismatch'] = (df['Header_Location'] != df['Predicted_Location'])

# Calculate Discrepancy Score
# Probability of TRUE header location
true_class_probs = []

for i in range(len(df)):
    true_label = y_encoded[i]
    class_index = np.where(model.classes_ == true_label)[0][0]
    prob = y_prob_full[i][class_index]
    true_class_probs.append(prob)

df['Discrepancy_Score'] = 1 - np.array(true_class_probs)


print(df[['Header_Location',
          'Predicted_Location',
          'Source_Mismatch',
          'Discrepancy_Score']].head())
```
